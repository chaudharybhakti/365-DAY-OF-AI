```python
import pandas as pd

data = {'Gender': ['M', 'Male', 'F', 'Female']}
df = pd.DataFrame(data)

df['Gender'] = df['Gender'].replace({'M': 'Male', 'F': 'Female'})
print(df)
```

```python
import pandas as pd
import numpy as np

data = {'Name': ['Amit', 'Sara'], 'Notes': [np.nan, np.nan]}
df = pd.DataFrame(data)

df = df.drop(columns=['Notes'])
print(df)
```

```python
import pandas as pd

data = {'Email': ['ABC@Gmail.com', 'XYZ@Yahoo.com']}
df = pd.DataFrame(data)

df['Email'] = df['Email'].str.lower()
print(df)
```

```python
import pandas as pd
import numpy as np

data = {'Name': ['Ravi', 'Sara', np.nan], 'Age': [25, np.nan, 30]}
df = pd.DataFrame(data)

df = df.dropna()
print(df)
```

```python
import pandas as pd

data = {'Date': ['2024-01-05', '05/01/2024', '2024-01-07']}
df = pd.DataFrame(data)

df['Date'] = pd.to_datetime(df['Date'], format='mixed')
print(df)
```

```python
import pandas as pd

data = {'Phone': ['987-654-3210', '(987)6543210']}
df = pd.DataFrame(data)

df['Phone'] = df['Phone'].str.replace(r'[^0-9]', '', regex=True)
print(df)
```

```python
import pandas as pd
import numpy as np

data = {'City': ['Pune', np.nan, 'Delhi']}
df = pd.DataFrame(data)

df['City'] = df['City'].fillna('Unknown')
print(df)
```
