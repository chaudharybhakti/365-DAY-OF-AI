```python
import pandas as pd
import numpy as np

# Messy data written directly in the code
data = {
    'Name': [' Rahul', 'Sneha ', 'RAHUL', 'Amit', None, 'sneha'],
    'Age': [25, 30, 25, np.nan, 28, 30],
    'City': ['Pune', 'mumbai', 'Pune ', 'Delhi', 'Mumbai', 'MUMBAI'],
    'Salary': [50000, 60000, 50000, 55000, None, 60000]
}

df = pd.DataFrame(data)

print("---- Original Data ----")
print(df)

# 1. Remove extra spaces and fix case in Name
df['Name'] = df['Name'].str.strip().str.title()

# 2. Fix inconsistent City names (spaces + case)
df['City'] = df['City'].str.strip().str.title()

# 3. Fill missing Age with average age
df['Age'] = df['Age'].fillna(df['Age'].mean())

# 4. Fill missing Salary with average salary
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())

# 5. Fill missing Name (if any) with "Unknown"
df['Name'] = df['Name'].fillna("Unknown")

# 6. Remove duplicate rows (same person appearing twice)
df = df.drop_duplicates(subset=['Name', 'Age', 'City'])

# 7. Reset index after cleaning
df = df.reset_index(drop=True)

print("\n---- Cleaned Data ----")
print(df)
```

```python
import pandas as pd
import numpy as np

data = {
    'Product': ['Pen', 'Pencil', 'PEN', 'Eraser', 'pencil', 'Notebook'],
    'Price': ['10', '5', '10', 'unknown', '5', '-20'],
    'Quantity': [100, 200, 100, 50, 200, 30],
    'Date': ['2024-01-05', '05/01/2024', '2024-01-05', '2024-01-06', '05/01/2024', '2024-01-07']
}

df = pd.DataFrame(data)
print("---- Original ----")
print(df)

# 1. Standardize product names
df['Product'] = df['Product'].str.strip().str.title()

# 2. Convert Price to numeric, invalid entries ('unknown') become NaN
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# 3. Remove negative prices (invalid data)
df = df[df['Price'] >= 0]

# 4. Fill missing prices with median
df['Price'] = df['Price'].fillna(df['Price'].median())

# 5. Standardize date formats
df['Date'] = pd.to_datetime(df['Date'], format='mixed')

# 6. Remove duplicates
df = df.drop_duplicates(subset=['Product', 'Date'])

df = df.reset_index(drop=True)
print("\n---- Cleaned ----")
print(df)
```

```python
import pandas as pd
import numpy as np

data = {
    'Student': ['Aman', 'Priya', 'aman', 'Rohit', 'Priya ', 'Kabir'],
    'Marks': [85, 150, 85, -5, 90, np.nan],   # 150 and -5 are invalid (marks should be 0-100)
    'Subject': ['Math', 'Science', 'Math', 'Math', 'science', 'English']
}

df = pd.DataFrame(data)
print("---- Original ----")
print(df)

# 1. Clean names
df['Student'] = df['Student'].str.strip().str.title()

# 2. Clean subject names
df['Subject'] = df['Subject'].str.strip().str.title()

# 3. Fix invalid marks (outside 0-100 range) -> set as NaN
df['Marks'] = df['Marks'].apply(lambda x: x if 0 <= x <= 100 else np.nan)

# 4. Fill missing marks with subject-wise average
df['Marks'] = df.groupby('Subject')['Marks'].transform(lambda x: x.fillna(x.mean()))

# 5. Remove duplicates
df = df.drop_duplicates(subset=['Student', 'Subject'])

df = df.reset_index(drop=True)
print("\n---- Cleaned ----")
print(df)
```

```python
import pandas as pd

data = {
    'Emp_Name': ['John Doe', 'jane smith', 'JOHN DOE', 'Amit Kumar'],
    'Email': ['John@Gmail.com', 'JANE@yahoo.com', 'john@gmail.com', 'amit@ gmail.com'],
    'Phone': ['9876543210', '98765-43210', '9876543210', '98765 43210']
}

df = pd.DataFrame(data)
print("---- Original ----")
print(df)

# 1. Standardize names
df['Emp_Name'] = df['Emp_Name'].str.strip().str.title()

# 2. Clean emails - lowercase and remove spaces
df['Email'] = df['Email'].str.strip().str.lower().str.replace(' ', '', regex=False)

# 3. Clean phone numbers - remove dashes/spaces, keep only digits
df['Phone'] = df['Phone'].str.replace(r'[^0-9]', '', regex=True)

# 4. Remove duplicates based on cleaned Email
df = df.drop_duplicates(subset=['Email'])

df = df.reset_index(drop=True)
print("\n---- Cleaned ----")
print(df)
```
