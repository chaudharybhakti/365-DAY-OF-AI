```python
import pandas as pd

data = {'Status': ['Active', 'Inactive', 'Test', 'Active']}
df = pd.DataFrame(data)

df = df[df['Status'] != 'Test']
print(df)
```

```python
import pandas as pd

data = {'Score': [95, 150, -20, 80]}
df = pd.DataFrame(data)

df['Score'] = df['Score'].clip(lower=0, upper=100)
print(df)
```

```python
import pandas as pd

data = {'Married': ['Yes', 'No', 'Yes', 'No']}
df = pd.DataFrame(data)

df['Married'] = df['Married'].map({'Yes': 1, 'No': 0})
print(df)
```

```python
import pandas as pd

data = {'Weight': ['70kg', '65 kg', '80KG']}
df = pd.DataFrame(data)

df['Weight'] = df['Weight'].str.extract(r'(\d+)').astype(int)
print(df)
```

```python
import pandas as pd

data = {'Name': ['Ravi', 'Ravi', 'Sara'], 'Age': [25, 25, 30]}
df = pd.DataFrame(data)

df = df.drop_duplicates(keep='first')
print(df)
```

```python
import pandas as pd
import numpy as np

data = {'Temp': [30, np.nan, np.nan, 32]}
df = pd.DataFrame(data)

df['Temp'] = df['Temp'].ffill()
print(df)
```
